```python
import pandas as pd


url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/SuicideChina.csv'
df = pd.read_csv(url)


print(df.head())


print(f"Dataset Dimensions: {df.shape}")
```

```python

print(df.describe())


print(df.info())
```

```python
print(df.isnull().sum())
df = df.dropna()
df = df.drop_duplicates()
print(f"Remaining data after cleaning: {df.shape}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(figsize=(10, 10), bins=20)
plt.show()

sns.countplot(data=df, x='Sex')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Died')
plt.title('Gender vs. Died')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x='Age', y='Died')
plt.title('Age vs. Died')
plt.xlabel('Age')
plt.ylabel('Died (yes=1, no=0)')
plt.show()

```

```python
sns.countplot(data=df, x='method', hue='Died')
plt.title('Method vs. Died')
plt.xlabel('Method of Suicide')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df = pd.get_dummies(df, drop_first=True)
print(df.head())
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, hue='Died')
plt.title('Pair Plot with Died as Hue')
plt.show()
```
